```python

import requests
from bs4 import BeautifulSoup

# Base URL to scrape
base_url = 'https://www.solotodo.cl/cells'

# List to hold all cellphones
cellphone_3 = []

# Loop through all 17 pages
for page in range(1, 18):
    # Construct the URL for the current page
    url = f"{base_url}?page_size=20&page={page}"

    # Send a GET request to the webpage
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract the data
        for card in soup.find_all('div', class_='MuiGrid-root MuiGrid-item css-1wxaqej'):
            name_elem = card.find('div', class_='MuiTypography-root MuiTypography-h5 css-11o89on')
            price_elem = card.find('div', class_='MuiTypography-root MuiTypography-h2 css-1lytq14')
            specs_elem = card.find('div', class_='ProductPage_product_specs__pYVnc')

            # Find product link
            product_div = card.find('div', class_='css-1g4yje1')
            product_link = product_div.find('a')['href'] if product_div and product_div.find('a') else None  # Extract href

            if name_elem and price_elem:
                name = name_elem.text.strip()
                price = price_elem.text.strip()

                spec_details = {}
                if specs_elem:
                    for dt in specs_elem.find_all('dt'):
                        key = dt.text.strip()
                        value = dt.find_next_sibling('dd').text.strip()
                        spec_details[key] = value

                # Create a unique identifier
                cellphone_entry = {
                    'Nombre': name,
                    'Precio': price,
                    'Link': product_link,  # Add product link to entry
                    **spec_details
                }

                # Check for duplicates before appending
                if cellphone_entry not in cellphone_3:
                    cellphone_3.append(cellphone_entry)

        print(f"Retrieved data from page {page}")
    else:
        print(f"Failed to retrieve data from {url}: {response.status_code}")

# Display the total number of cellphones retrieved
print(f"Total cellphones retrieved: {len(cellphone_3)}")
# Optionally, display all retrieved cellphones
for cellphone in cellphone_3:
    print(cellphone) 
```

```python
# Transform the cellphones list into a Pandas DataFrame
import pandas as pd

df3 = pd.DataFrame(cellphone_3)

# Display the DataFrame
print(f"Total cellphones retrieved: {len(df3)}")
print(df3.head())  # Display the first few rows of the DataFrame"
```

```python
# Create a list of links from the 'Link' column
link_list = df3['Link'].tolist()

# Print the list
print(link_list)
```
